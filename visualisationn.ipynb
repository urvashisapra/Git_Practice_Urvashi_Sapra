```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/urvashisapra/Module4/main/BAC.csv'
data = pd.read_csv(url)
data.head()
```

```python
print(data.columns)
```

```python
# Line Chart - Example
plt.figure(figsize=(10, 5))
plt.plot(data['Adj Close'].nlargest(20), marker='o', linestyle='-', color='b')
plt.title('Top 20 Companies by Market Cap')
plt.xlabel('Open')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
# Bar Chart - Example
data_sorted = data.nlargest(10, 'Volume')
sns.barplot(x='Close', y='Volume', data=data_sorted, palette='viridis')
plt.xticks(rotation=90)
plt.title('Top 10 Companies by Market Cap')
plt.show()
```

```python
# Scatter Plot - Example
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Open'], y=data['High'], hue=data['Low'], alpha=0.7)
plt.title('Stock Price vs Dividend Yield')
plt.xlabel('Stock Price')
plt.ylabel('Dividend Yield')
plt.show()
```

```python
# Histogram - Example
plt.figure(figsize=(8, 6))
plt.hist(data['Volume'], bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Stock Prices')
plt.xlabel('Open')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot - Example
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Close'], y=data['Volume'])
plt.xticks(rotation=90)
plt.title('Market Cap Distribution by Sector')
plt.show()
```

```python
# Interactive Plot using Plotly
fig = px.scatter(data, x='Volume', y='Open', color='Close', size='Adj Close', hover_data=['Low'])
fig.show()
```
